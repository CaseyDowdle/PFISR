```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import numpy as np
import scipy
```

```python
theta = 2*np.pi*np.linspace(0,1,100) # azimuth of wave propagation
T = 16.4*60 # period of wave in 1/sec
w = 2*np.pi/T # angular frequency
t = 1 # time
#z = np.cos(theta) # height
z = 0
U_h = 2.84 # amplitude of wave, for now define to be U_h given by plot
H = 50. # neutral atmospheric scale height in km
k_x =  2*np.pi/50.
k_z =  -2*np.pi/100.

k = np.array([k_x,k_z])
r = 10*np.array([np.cos(theta),np.sin(theta)])
I = 55./360. # magnetic inclination in radians

u_p = np.exp(-z/(2*H))*np.cos(w*t-np.dot(k,r)) # GW phase
U_z = U_h*np.tan(I) 
u = np.array([U_h*u_p,U_z*u_p])

B = np.array([np.cos(I),np.sin(I)]) # geomagnetic field unit vector
ion_grad = 0

Ne_p = np.zeros(100)
for i in range(100):
    Ne_p[i] = (np.dot(u[:,i],B)/w)*(np.dot(k,B)*(1./(2*H) + ion_grad))
```

```python
ax = plt.subplot(111,projection = 'polar')
ax.plot(theta,Ne_p)
```
